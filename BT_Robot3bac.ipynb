```python
import math as m
import numpy as np
import csv
mo_phong=[]
for i in range(0,180,1):
  arr=[] 
  for j in range(0,180,1):
      for k in range(0,180,1):
        x=50*m.cos(m.radians(i))+40*m.cos(m.radians(i+j))+30*m.cos(m.radians(i+j+k))
        y=50*m.sin(m.radians(i))+40*m.sin(m.radians(i+j+k))+30*m.sin(m.radians(i+j+k))
        si=i+j+k
        arr.append(np.array((i,j,k,x,y,si)))
  mo_phong.extend(arr)
```

```python
header = ['tt1', 'tt2','tt3', 'Px', 'Py','si']
with open('Robot3bac.csv', 'w', encoding='UTF8', newline='') as data_table:
    writer = csv.writer(data_table)
    writer.writerow(header)
    writer.writerows(mo_phong)
```

```python
import pandas as pd
data = pd.read_csv ('Robot3bac.csv')
data.head()
```

```python
x=data.iloc[:,[3,4,5]]
y=data.iloc[:,[0,1,2]]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping

model=Sequential()
model.add(Dense(128, activation='relu', input_shape = (3,))) 
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3,  activation='linear'))
model.summary()
```

```python
model.compile(loss='mse',metrics=['mean_absolute_error'])
history= model.fit(x_train, y_train, batch_size=100, epochs=50, verbose=1, validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss', patience=50)])
```

```python
score=model.evaluate(x_test,y_test,verbose=0)
print('Test loss',score[0])
print('Test accuracy',score[1])
```

```python
predictions=model.predict(x_test)
y_test=np.array(y_test)
print("gia tri du doan:", predictions[1])
print("gia tri thuc:",y_test[1])
```
